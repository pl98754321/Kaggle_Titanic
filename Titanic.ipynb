```python
def nomailize(data):
  return (data - data.min())/(data.max()-data.min())

def Check_Catagory(index,check = "Survived"): #Mean of Check when given index
  return train[[index,check]].groupby([index], as_index=False).mean().sort_values(by=check, ascending=False)

def detect_outliers(df,features):
    outlier_All = []
    multi_outlier = []
    n = 2
    
    for col in features:
        Q1 = np.percentile(df[col], 25)
        Q3 = np.percentile(df[col],75)
        IQR = Q3 - Q1
        outlier_step = 1.5 * IQR
        outlier_list_col = df[(df[col] < Q1 - outlier_step) | (df[col] > Q3 + outlier_step )].index
        outlier_All.extend(outlier_list_col)

    outlier_indices = Counter(outlier_All)
    multi_outlier = list(k for k,v in outlier_indices.items() if v>n)
    return multi_outlier
```

```python
data = pd.read_csv("/content/drive/MyDrive/Colab/Kaggle_titanic/train.csv")
data_test = pd.read_csv("/content/drive/MyDrive/Colab/Kaggle_titanic/test.csv")
```

```python
Feature = ["Pclass","Sex","Embarked","SibSp","Parch"]
train = pd.get_dummies(data[Feature])
test = pd.get_dummies(data_test[Feature])
Label = data["Survived"]
```

```python
Xt , Xv , Yt , Yv = train_test_split(train,Label,train_size=0.9)
```

```python
model = RandomForestClassifier()
model.fit(Xt,Yt)
```
