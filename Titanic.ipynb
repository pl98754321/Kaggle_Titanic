```python
import numpy as np
import pandas as pd
from tensorflow.keras import layers,models
import matplotlib.pyplot as plt
from tensorflow import keras
```

```python
def nomailize(data):
  return (data - data.min())/(data.max()-data.min())
```

```python
data = pd.read_csv("/content/drive/MyDrive/Colab/Kaggle_titanic/train.csv")
```

```python
features = ["Pclass", "Sex", "SibSp", "Parch","Age","Fare"]
Xt = pd.get_dummies(data[features])
Yt = data['Survived']
```

```python
Xt['Age'] = Xt["Age"].fillna(Xt["Age"].mean())
```

```python
Columns = Xt.columns
Xt[Columns] = nomailize(Xt[Columns])
```

```python
nxt = np.array(Xt).reshape(-1,7,1)
nyt = np.array(Yt)
```

```python
model = keras.Sequential()
model.add(layers.Dense(48, input_shape = (7,),activation="relu"))
model.add(layers.Dense(24,activation="relu"))
model.add(layers.Dense(12,activation="relu"))
model.add(layers.Dense(6,activation="relu"))
model.add(layers.Dense(1,activation="relu"))
model.compile(loss=keras.losses.MeanSquaredError(),optimizer="sgd")
```

```python
model = keras.applications.DenseNet121()
```

```python
keras.utils.plot_model(model,show_shapes=True)
```

```python
model.fit(nxt,nyt,epochs=500)
```

```python
nzt = model.predict(nxt)
nzt = nzt.argmax(axis=1)
```

```python
np.sum(nzt == nyt.argmax(axis=1))/(len(nyt))
```
